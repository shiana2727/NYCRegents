From the data it is clear when comparing Stuyvesant High School and Marta Valle High School, that Marta Valle High School scores significantly lower than Stuyvesant High School. On average students in Marta Valle High School scored lower in their algebra 2 trig regents exam than the other regents exams. In Stuyvesant students scored lower in Common Core Algebra, however majority passed with a grade over an 80. Stuyvesant is a rigorous and an academically challenging high school, that takes their students based on an entrance exam whereas Marte Valle was a zoned school and took students based on the vicinity of where they live from the school.
